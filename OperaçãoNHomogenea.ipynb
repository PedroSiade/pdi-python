Operações homogêneas e não homogêneas:


Operações homogêneas são feitas em cada pixel, sem considerar a vizinhanhança. 

---

Um exemplo de operação homogênea é fazer o negativo de uma imagem, pois para cada ponto f(x,y) é transformado no inverso na nova matriz. 

256-f(x,y)= h(x,y)
Sendo assim se f(1,2)=10, na imagem negativada vai ser 246, sendo o processo repetido para cada ponto.

A operação não homogênea é aquela em que a transformação depende de informações externas à vizinhança de cada pixel, sendo que regiões diferentes serão processadas por funções diferentes. Um exemplo de operação não homogênea é a restauração de imagens/photoshops, utiliza-se a região vizinha a fim de corrigir a região com problema.

```python
# importa modulos OpenCV e Matplotlib
import cv2
import matplotlib.pyplot as plt
import numpy as np   
```

```python
imagem = cv2.imread('imagens_AT3/Fig0305(a)(DFT_no_log).tif', cv2.IMREAD_GRAYSCALE).astype(float)
fxy=imagem
plt.imshow(fxy, cmap='gray')
```

```python
c=1.0
gxy = (c*np.log(1+fxy))
plt.imshow(gxy, cmap='gray')
```

Na imagem acima, é possível observar o efeito da transformação logaritmica, uma transformação homogênea a qual expande os valores de pixels mais escuro, no entanto comprimindo as intensidades mais alta. Como resultado, a faixa dinâmica é comprimida, diminuindo a diferença entre a intensidade mais alta e baixa e realçando os pixels de menor valor.

Dessa forma, torna-se possível visualizar com maior clareza os raios lunares e as pequenas variações que eles causam na imagem, potencializando os efeitos desses raios na fotografia.



```python
imagem = cv2.imread('imagens_AT3/Captura de tela 2023-05-31 133736.png')
gxy = c*fxy**0.32
fig, eixo = plt.subplots(1,2, figsize=(20,10))
eixo[0].imshow(imagem)
eixo[1].imshow(gxy, cmap='gray')




```

Na etapa acima foi realizado um comparativo por meio do gráfico entre as funções, testando-se vários valores para gama. Sendo assim, o valor de gama da transformação por meio de potencia que mais se aproximou da outra transformação foi 0.32. 



```python
gamas = [3.0, 0.6, 0.3, 0.1]
gxy = [c*fxy**gama for gama in gamas]
fig = plt.figure(figsize=(15,15))
plt.title("")
plt.axis('off')
axeslist = fig.subplots(2,2)
for (i, axes) in enumerate(axeslist.ravel()):
  axes.imshow(gxy[i],cmap='gray')
  axes.set_title(f"$\gamma={gamas[i]}$")
  axes.set_axis_off()
plt.show()
```


Foram realizadas várias transformações com diferentes valores de gama. Observa-se que, quando o gama é maior que 1, as regiões mais claras são menos realçadas, enquanto um gama menor resulta em um realce maior dos raios.
