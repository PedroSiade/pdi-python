Formulas: L = 2^k e b = M*N*k

Foi possvel obter os mesmos resultados observados no livro nos exemplos fazendo a requantizao da imagem acima. Percebe-se que a matriz(numero de coluna e linhas) se mantm igual  matriz da imagem original, no entanto altera-se a quantidade de cinzas diferentes. 

Sendo assim, haver uma diminuio do nmero de bits para armazenar a fotografia, diminuindo o armazenamento necessrio. Exemplo:

Imagem 256 bits:

L=2^8=256

b=(452*374*8)/1024=1320,69

Imagem de 1 bit:

L=2^1=2

b=452*374*1/1024=165,09

Dessa forma, percebe-se que a diminuio da faixa dinmica/ nmero de intensidade diferentes ocasiona a reduo da quantidade de armazenamento utilizado. Logo, pode-se afirmar que dependendo da necessidade do usurio  possvel reduzir a quantidade de armazenamento da imagem, mas sempre analisando a perca da qualidade observada. 

Isto pode ser exemplificado a partir da terceira requantizao, pois surgem faixas/sulcros no crnio, por conta da mudana abrupta de cinza(falta de intensidades intermedirias), dificultando a anlise do raio x.
