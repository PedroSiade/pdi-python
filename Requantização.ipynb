Fazendo a requantização da imagem do livro e exibindo quantidade de tons de cinza diferentes em cada imagem requantizada:

```python
bit = 8
requantizada_imagem1 = (imagem2 / (2**(8 - bit))).astype('uint8')
print('Numero de valores unicos : ', len(np.unique(requantizada_imagem1)))
bit = 7
requantizada_imagem2 = (imagem2 / (2**(8 - bit))).astype('uint8')
print('Numero de valores unicos requantização 1: ', len(np.unique(requantizada_imagem2)))
bit = 6
requantizada_imagem3 = (imagem2 / (2**(8 - bit))).astype('uint8')
print('Numero de valores unicos requantização 2: ', len(np.unique(requantizada_imagem3)))
bit = 4
requantizada_imagem4 = (imagem2 / (2**(8- bit))).astype('uint8')
print('Numero de valores unicos requantização 3: ', len(np.unique(requantizada_imagem4)))
bit = 2
requantizada_imagem5 = (imagem2 / (2**(8 - bit))).astype('uint8')
print('Numero de valores unicos requantização 4: ', len(np.unique(requantizada_imagem5)))
bit = 1
requantizada_imagem6 = (imagem2 / (2**(8 - bit))).astype('uint8')
print('Numero de valores unicos requantização 5: ', len(np.unique(requantizada_imagem6)))
print(requantizada_imagem6.shape)
print(requantizada_imagem6.dtype)

fig, eixo = plt.subplots(2, 3, figsize=(50,20))

eixo[0,0].imshow(requantizada_imagem1, cmap='gray')
eixo[0,1].imshow(requantizada_imagem2, cmap='gray')
eixo[0,2].imshow(requantizada_imagem3, cmap='gray')
eixo[1,0].imshow(requantizada_imagem4, cmap='gray')
eixo[1,1].imshow(requantizada_imagem5, cmap='gray')
eixo[1,2].imshow(requantizada_imagem6, cmap='gray')
```

Formulas: L = 2^k e b = M*N*k

Foi possível obter os mesmos resultados observados no livro nos exemplos fazendo a requantização da imagem acima. Percebe-se que a matriz(numero de coluna e linhas) se mantém igual à matriz da imagem original, no entanto altera-se a quantidade de cinzas diferentes. 

Sendo assim, haverá uma diminuição do número de bits para armazenar a fotografia, diminuindo o armazenamento necessário. Exemplo:

Imagem 256 bits:

L=2^8=256

b=(452*374*8)/1024=1320,69

Imagem de 1 bit:

L=2^1=2

b=452*374*1/1024=165,09

Dessa forma, percebe-se que a diminuição da faixa dinâmica/ número de intensidade diferentes ocasiona a redução da quantidade de armazenamento utilizado. Logo, pode-se afirmar que dependendo da necessidade do usuário é possível reduzir a quantidade de armazenamento da imagem, mas sempre analisando a perca da qualidade observada. 

Isto pode ser exemplificado a partir da terceira requantização, pois surgem faixas/sulcros no crânio, por conta da mudança abrupta de cinza(falta de intensidades intermediárias), dificultando a análise do raio x. 

Fazendo a requantização da imagem do raio x fornecidade, e exibindo quantidade de tons de cinza diferentes em cada imagem requantizada:

```python
bit = 11
requantizada_imagem1 = (imagem / (2**(11 - bit))).astype('uint16')
print('Numero de valores unicos : ', len(np.unique(requantizada_imagem1)))
bit = 8
requantizada_imagem2 = (imagem / (2**(11 - bit))).astype('uint16')
print('Numero de valores unicos requantização 1: ', len(np.unique(requantizada_imagem2)))
bit = 4
requantizada_imagem3 = (imagem / (2**(11 - bit))).astype('uint16')
print('Numero de valores unicos requantização 2: ', len(np.unique(requantizada_imagem3)))
bit = 2
requantizada_imagem4 = (imagem / (2**(11- bit))).astype('uint16')
print('Numero de valores unicos requantização 3: ', len(np.unique(requantizada_imagem4)))
bit = 1
requantizada_imagem5 = (imagem / (2**(11 - bit))).astype('uint16')
print('Numero de valores unicos requantização 4: ', len(np.unique(requantizada_imagem5)))
bit = 0
requantizada_imagem6 = (imagem / (2**(11 - bit))).astype('uint16')
print('Numero de valores unicos requantização 5: ', len(np.unique(requantizada_imagem6)))
print(requantizada_imagem6.shape)
print(requantizada_imagem6.dtype)

fig, eixo = plt.subplots(2, 3, figsize=(50,20))

eixo[0,0].imshow(requantizada_imagem1, cmap='gray')
eixo[0,1].imshow(requantizada_imagem2, cmap='gray')
eixo[0,2].imshow(requantizada_imagem3, cmap='gray')
eixo[1,0].imshow(requantizada_imagem4, cmap='gray')
eixo[1,1].imshow(requantizada_imagem5, cmap='gray')
eixo[1,2].imshow(requantizada_imagem6, cmap='gray')

```
